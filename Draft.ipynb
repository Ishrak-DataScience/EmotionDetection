```python
# prompt: I want to upload all csv flies in a folder

from google.colab import files
import pandas as pd
import io
import os

# Upload all CSV files from a local folder
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  if fn.endswith('.csv'):
    try:
      df = pd.read_csv(io.BytesIO(uploaded[fn]))

      # Display basic info about the DataFrame
      print(f"DataFrame info for {fn}:")
      print(df.info())

      # Display the first few rows
      print(f"First few rows of {fn}:")
      print(df.head())

      # Display descriptive statistics
      print(f"Descriptive statistics for {fn}:")
      print(df.describe())

      # Check for missing values
      print(f"Missing values in {fn}:")
      print(df.isnull().sum())

      # Explore unique values in columns
      for column in df.columns:
        print(f"Unique values in {column} of {fn}: {df[column].unique()}")

    except Exception as e:
      print(f"Error processing {fn}: {e}")
```
